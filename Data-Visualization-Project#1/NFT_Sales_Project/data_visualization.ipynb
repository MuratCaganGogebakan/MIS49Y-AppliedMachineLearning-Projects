```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data

df = pd.read_csv('/Users/yunsemr/Downloads/boun-ders/2022-Spring/MIS-49Y/HW1/NFT_Sales.csv')

df.head()
```

```python
df.shape
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
#getting rid of NaN's

df['Active_Market_Wallets_cumsum'] = df['Active_Market_Wallets_cumsum'].fillna(df['Active_Market_Wallets_cumsum'].mean())
df['Active_Market_Wallets'] = df['Active_Market_Wallets'].fillna(df['Active_Market_Wallets'].mean())
df.loc[0, 'Active_Market_Wallets'] = 0 #Assigning the first element 0 in order to mean value
df = df.fillna(0)
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Data visualization

plt.figure(figsize=(16,5), dpi=100)
plt.plot(df.index, df.Number_of_Sales, color='tab:blue')
plt.gca().set(title='Daily NFT Sales.', xlabel='Date', ylabel='NFT')
plt.show()

```

```python

sns.lineplot(data=df, x= df.Date, y = "Number_of_Sales" )
plt.show()
```

```python
sns.lineplot(data=df, x="Date", y="Sales_USD")
plt.show()
```

```python
sns.kdeplot(data=df, y=df.index, x="Sales_USD", fill= True)
plt.show()
```
